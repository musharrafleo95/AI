We do pair bootstrap when we want to say something the data underlying a model. For example we collected voting data from a sample and and made a regression model for it. But that model is just for a sample and our requirement is a model that shows the linear relationship for the whole populations so this is where bootstrap jumps in. It is called pair bootstrap because we have two variables to do bootstrap on. So to achieve that we do bootstrap on the indices of data

```python

def bs_pairs_linreg(x, y, size = 1):
    '''perform pair bootstrap for linear regression'''
    #getting the indices
    inds = np.arange(len(x)) #as x and y have same length
    #making empty variables
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)
    #generating bootstrap slop and intercepts
    for i in range(size):
        bs_ind = np.random.choice(inds, size = len(inds))
        bs_x, bs_y = x[bs_ind], y[bs_ind]
        bs_slope_reps[i], bs_intercept_reps[i] = np.polyfit(bs_x, bs_y, 1)
    
    return bs_slope_reps, bs_intercept_reps
    
```

```python
np.random.seed(41)
#Performing pair bootstrap on illiteracy and fertility rate.
bs_slope_reps, bs_intercept_reps = bs_pairs_linreg(illiteracy, fertility, size = 10000)

#Computing the 95% confidence interval for slopes
print(np.percentile(bs_slope_reps, [2.5, 97.5]))

#plotting the histogram
_ = plt.hist(bs_slope_reps, bins = 50, density = True)
_ = plt.xlabel('Slope')
_ = plt.ylabel('PDF')
plt.show()
```

## Plotting bootstrap Regressions
To see the variability we might expect in the linear regression we will plot the the line from each bootstrap replicate. We will plot the 1000 values so that our plot dont get messy

```python
x = np.array([0, 100])
#plotting the regression lines
for i in range(1000):
    _ = plt.plot(x, x*bs_slope_reps[i] + bs_intercept_reps[i], alpha = 0.02, linewidth = 0.5, color = 'red')
    
_ = plt.plot(illiteracy, fertility, marker = '.', linestyle = 'none')
_ = plt.xlabel('Illiteracy')
_ = plt.ylabel('Fertility')
plt.margins(0.02)
plt.show()
```
