It is very important to do exploratory data analysis before doing estimation or fitting any model. This statement can be prove by Anscombe Dataset. </p>
Although the four dataset have same slope and intercept but they all are too different from each other.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
anscombe_x = np.array([[10.,  8., 13.,  9., 11., 14.,  6.,  4., 12.,  7.,  5.],
                    [10.,  8., 13.,  9., 11., 14.,  6.,  4., 12.,  7.,  5.],
                    [10.,  8., 13.,  9., 11., 14.,  6.,  4., 12.,  7.,  5.],
                    [ 8.,  8.,  8.,  8.,  8.,  8.,  8., 19.,  8.,  8.,  8.]])
anscombe_y = np.array([[ 8.04,  6.95,  7.58,  8.81,  8.33,  9.96,  7.24,  4.26, 10.84,4.82,  5.68],
                     [9.14, 8.14, 8.74, 8.77, 9.26, 8.1 , 6.13, 3.1 , 9.13, 7.26, 4.74],
                     [ 7.46,  6.77, 12.74,  7.11,  7.81,  8.84,  6.08,  5.39,  8.15, 6.42,  5.73],
                     [ 6.58,  5.76,  7.71,  8.84,  8.47,  7.04,  5.25, 12.5 ,  5.56,7.91,  6.89]])
```

```python
#plottinh all the figure
plt.figure(figsize = (20,10))
i = 1
for x,y in zip(anscombe_x, anscombe_y):
    slope, intercept = np.polyfit(x, y, 1)
    print('slope:', slope, 'intercept:', intercept)
    plt.subplot(2,2,i)
    _ = plt.plot(x, y, marker = '.', linestyle = 'none', color = 'blue')
    x_line = np.array([0, max(x)+2])
    y_line = slope*x_line + intercept
    _ = plt.plot(x_line, y_line, color = 'black')
    
    i +=1
```

here we can see that although the scatter plot are different but we have the same regression line which can also be prove by the almost same slope and intercept

SO therefore before doing any modeling we must first visualize the data.
