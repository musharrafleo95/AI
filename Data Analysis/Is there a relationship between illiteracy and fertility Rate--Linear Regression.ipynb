```python
def pearson_r(x,y):
    coef = np.corrcoef(x, y)
    return coef[0,1]
```

```python
#plotting fertility and illiteracy to do eda
_ = plt.plot(illiteracy, fertility, marker = '.', linestyle = 'none')
plt.margins(0.02)
_ = plt.xlabel(' Percent illiterate')
_ = plt.ylabel('Fertility')
plt.show()
```

As we can see there is a linear relationship between illiteracy and fertility. We can further confirm this by generating Pearson correlation coefficient

```python
pearson_r(illiteracy, fertility)
```

Since there is a linear relationship present so we will try to qauntify this relationship with a linear model. Here I will use linear regression.

```python
# Performing a linear regression .
slope, intercept = np.polyfit(illiteracy, fertility, 1)

#Making a regression line in plot
x = np.array([0, 100])
y = slope*x + intercept

#plotting fertility and illiteracy with regression line
_ = plt.plot(illiteracy, fertility, marker = '.', linestyle = 'none')
_ = plt.plot(x, y)
plt.margins(0.02)
_ = plt.xlabel(' Percent illiterate')
_ = plt.ylabel('Fertility')
plt.show()
```

SO here we can see that we quantifed our relation ship between illiteracy and fertility by the equation
<b>y = slope*x + intercept</b>, where slope and intercept are the value we got from np.polyfit()
